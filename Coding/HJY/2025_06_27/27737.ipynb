```python
from collections import deque

inputs = input().strip()
n, m, k = map(int, inputs.split())
matrix = []
for _ in range(n):
    parts = input().strip().split()
    matrix.append([int(x) for x in parts])

def check(matrix):
    for i in range(n):
        for j in range(n):
            if matrix[i][j] == 0:
                return False
    return True

queue = deque([])
arr = []
temp = 0

while not check(matrix):
    if 0 < len(queue): 
        (x,y) = queue.popleft()
        if x > 0 and matrix[x-1][y] == 0:
            temp += 1
            queue.append((x-1, y))
            matrix[x-1][y] = 2
        if x < n-1 and matrix[x+1][y] == 0:
            temp += 1
            queue.append((x+1, y))
            matrix[x+1][y] = 2
        if y > 0 and matrix[x][y-1] == 0:
            temp += 1
            queue.append((x, y-1))
            matrix[x][y-1] = 2
        if y < n-1 and matrix[x][y+1] == 0:
            temp += 1
            queue.append((x, y+1))
            matrix[x][y+1] = 2
    else:
        if temp > 0:
            arr.append(temp)
        temp = 0
        for i in range(n):
            for j in range(n):
                if matrix[i][j] == 0:
                    queue.append((i, j))
                    matrix[i][j] = 2
                    temp += 1
                    break
            if temp > 0:
                break
            
if temp > 0:
    arr.append(temp)
    
if len(arr) == 0:
    print("IMPOSSIBLE")

elif k == 1 :
    if sum(arr) <= m:
        print("POSSIBLE")
        print(m-sum(arr))
    else:
        print("IMPOSSIBLE")
elif k > 1:
    for i in range(len(arr)):
        if arr[i] % k != 0:
            arr[i] = arr[i] // k + 1
        else:
            arr[i] = arr[i] // k
    if sum(arr) <= m:
        print("POSSIBLE")
        print(m - sum(arr))
    else:
        print("IMPOSSIBLE")
```
