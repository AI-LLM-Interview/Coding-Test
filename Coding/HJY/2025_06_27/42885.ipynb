무인도에 갇힌 사람들을 구명보트를 이용하여 구출하려고 합니다. 구명보트는 작아서 한 번에 최대 2명씩 밖에 탈 수 없고, 무게 제한도 있습니다.

예를 들어, 사람들의 몸무게가 [70kg, 50kg, 80kg, 50kg]이고 구명보트의 무게 제한이 100kg이라면 2번째 사람과 4번째 사람은 같이 탈 수 있지만 1번째 사람과 3번째 사람의 무게의 합은 150kg이므로 구명보트의 무게 제한을 초과하여 같이 탈 수 없습니다.

구명보트를 최대한 적게 사용하여 모든 사람을 구출하려고 합니다.

사람들의 몸무게를 담은 배열 people과 구명보트의 무게 제한 limit가 매개변수로 주어질 때, 모든 사람을 구출하기 위해 필요한 구명보트 개수의 최솟값을 return 하도록 solution 함수를 작성해주세요.

```python
def solution(people, limit):
    answer = 0
    people.sort()
    li = []
    temp = 0
    for p in people:
        if temp + p <= limit:
            temp += p
        else:
            li.append(temp)
            temp = 0
            temp += p
    if temp > 0:
        li.append(temp)
    answer = len(li)
    return answer        

solution([20, 30, 70, 80], 100)
# expect 2
```

```python
def solution(people, limit):
    answer = 0
    people.sort()
    li = []
    temp = 0
    while len(people) > 0:
        p = people.pop()
        if temp + p <= limit:
            temp += p
        else:
            while len(people) > 0:
                _p = people[0]
                if temp + _p <= limit:
                    temp += _p
                    people.pop(0)
                else:
                    break
            li.append(temp)
            temp = p
    if 0 < temp:
        li.append(temp)
    return len(li)

solution([20, 20, 20, 20], 100)
# expect 2
```

### 한 보트에 최대 2명씩 밖에 탈수 없다

```python
def solution(people, limit):
    answer = 0
    people.sort()
    li = []
    temp = []
    while len(people) > 0:
        p = people.pop()
        if len(temp) < 2 and sum(temp) + p <= limit:
            temp.append(p)
        else:
            while len(people) > 0:
                _p = people[0]
                if len(temp) < 2 and sum(temp) + _p <= limit:
                    temp.append(_p)
                    people.pop(0)
                else:
                    break
            li.append(temp)
            temp = []
            temp.append(p)
    if 0 < len(temp):
        li.append(temp)
    return len(li)


solution([20, 20, 20, 20], 100)
# expect 2
```

### 그러나 효율성 검사에서 탈락
-> deque 사용 -> 성공!

```python
from collections import deque

def solution(people, limit):
    answer = 0
    people.sort()
    people = deque(people)
    li = []
    temp = []
    while len(people) > 0:
        p = people.pop()
        if len(temp) < 2 and sum(temp) + p <= limit:
            temp.append(p)
        else:
            while len(people) > 0:
                _p = people[0]
                if len(temp) < 2 and sum(temp) + _p <= limit:
                    temp.append(_p)
                    people.popleft()
                else:
                    break
            li.append(temp)
            temp = []
            temp.append(p)
    if 0 < len(temp):
        li.append(temp)
    return len(li)
```

## 📌 왜 deque를 쓰는가?
파이썬의 기본 list는:

- pop() → O(1)

- pop(0) (또는 people.pop(0)) → O(n) ❌

반면, collections.deque는:

- pop() → O(1)

- popleft() → O(1) ✅

따라서 앞에서 사람을 꺼낼 일이 있다면 deque가 훨씬 효율적입니다.

# 지피티

```python
from collections import deque

def solution(people, limit):
    people.sort()
    dq = deque(people)
    answer = 0

    while dq:
        if len(dq) >= 2 and dq[0] + dq[-1] <= limit:
            dq.popleft()
        dq.pop()
        answer += 1

    return answer
```
